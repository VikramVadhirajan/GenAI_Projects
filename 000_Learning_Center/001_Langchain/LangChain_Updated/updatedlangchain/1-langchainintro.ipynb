```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
from langchain.agents import create_react_agent, AgentExecutor
from langchain_core.tools import tool
from langchain_openai import ChatOpenAI
from langchain import hub

@tool
def get_weather(city: str) -> str:
    """Get the weather for a city."""
    return f"The weather in {city} is windy."

# Load ReAct prompt
prompt = hub.pull("hwchase17/react")

# Create LLM
llm = ChatOpenAI(model="gpt-4o-mini")

# Create agent
agent = create_react_agent(llm, [get_weather], prompt)
agent
```

```python

# Agent executor
agent_executor = AgentExecutor(agent=agent, tools=[get_weather], verbose=True)

# Run
agent_executor.invoke({"input": "What is the weather in Paris?"})
```
